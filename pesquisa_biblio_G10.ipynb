# Pesquisa Bibliográfica
## Análise genética da resistência a antibióticos em Legionella pneumophila
### Mestrado em Bioinformática, Laboratórios de Bioinformática, Grupo 10

```python
from Bio import Entrez
from Bio import Medline
```

```python
#Palavras-chave da pesquisa
term1 = "gene antibiotic resistance Legionella pneumophila"
term2 = "LpeB"
term3 = 'adeF'
#term4 = 'Legionella pneumophila'
terms = [term1]
```

```python
for term in terms:
    print(f"----> Artigos relacionados a '{term}' <----\n")
    # --> procurar o número de artigos que contêm os termos indicados <--
    Entrez.email = ""
    handle = Entrez.egquery(term=term)
    record = Entrez.read(handle)
    for row in record["eGQueryResult"]:
        if row["DbName"] == "pubmed":
            num_artigos = row["Count"]
    handle.close()
    # --> listar os IDs dos artigos que contêm os termos indicados <--
    Entrez.email = ""
    handle = Entrez.esearch(db="pubmed",term=term,retmax=num_artigos)
    record = Entrez.read(handle)
    idlist = record["IdList"]
    handle.close()
    # --> listar o título, autores e source dos artigos encontrados anteriormente <--
    Entrez.email = ""
    handle = Entrez.efetch(db="pubmed",id=idlist,rettype="medline",retmode="text")
    record = Medline.parse(handle)
    records = list(record)
    for record in records:
        print("Title:", record.get("TI", "?"))
        print("Authors:", ", ".join(record.get("AU", "?")))
        print("Source:", record.get("SO", "?"))
        print("")
    print("")
    handle.close()
```
